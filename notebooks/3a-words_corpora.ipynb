```python
import re
def findWords(s,pattern):
    # \W[A-Z][a-z]*
    # ([A-Z]\w*)
    # r"([A-Z][a-z]+)"
    # r"\W([A-Z]\w*)\W"
    # r"\b([A-Z]\w*)\b"    (\b stands for boundary of word)
    #pattern = r"\b[A-Z][a-z]*\b"
    m = re.finditer(pattern, s)
    matches = list(m)
    return [x.group(0) for x in matches]
```

```python
with open('shakes.txt','r') as f:
    corpus = f.read()

len(corpus)
```

```python
corpus[0:100]
```

```python
size = len(findWords(corpus,r'\b\w+\b'))
size
```

```python
nI = len(re.findall(r'\bI\b',corpus))
nI
```

```python
# (this will miss some occurences)
nIam = len(re.findall(r'\bI am\b',corpus))
nIam
```

```python
# a little better
nIam = len(re.findall(r'\bI[ \n\r]+am\b',corpus))
nIam
```

```python
# P(am | I) = P(I am) / P (I)  => #(I am) / #(I)
P_cond_am_I = nIam / nI
P_cond_am_I
```

```python
# what is probability of I? P(I)
number_of_words = len(re.findall(r'\w+',corpus))
number_of_words
```

```python
P_I = nI / number_of_words
P_I
```
