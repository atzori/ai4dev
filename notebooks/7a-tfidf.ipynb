## TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Esempio di documenti
documents = [
    "Questo è il primo documento.",
    "Questo documento è il secondo documento.",
    "E questo è il terzo, il terzo documento."
]

# Inizializzazione del TfidfVectorizer
vectorizer = TfidfVectorizer(use_idf=True,smooth_idf=True, sublinear_tf=False)

# Adattamento del vettorizzatore ai documenti e trasformazione dei documenti in una rappresentazione TF-IDF
tfidf_matrix = vectorizer.fit_transform(documents)

# Visualizzazione delle feature (parole)
print("Feature (parole) estratte:")
print(vectorizer.get_feature_names_out())

# Visualizzazione della rappresentazione TF-IDF dei documenti
print("\nRappresentazione TF-IDF dei documenti:")
print(tfidf_matrix.toarray())
```
