````python
# let's try with these regex:
# [a-zA-Z]ing[^a-zA-Z]
# (\w*ing)\W
pattern = r"([a-zA-Z]+ing)[^a-zA-Z]"
m = re.finditer(pattern, s)
matches = list(m)
````

````python
import re
def findWords(s):
    # \W[A-Z][a-z]*
    # ([A-Z]\w*)
    # r"([A-Z][a-z]+)"
    # r"\W([A-Z]\w*)\W"
    # r"\b([A-Z]\w*)\b"    (\b stands for boundary of word)
    pattern = r"\b[A-Z][a-z]*\b"
    m = re.finditer(pattern, s)
    matches = list(m)
    return [x.group(0) for x in matches]
````

````python
findWords("This is a sentence with some Capitalized Words, and I nEEd to Find Them")
````

## Substitution of strings with regex

````python
re.sub(r'ing',r'<ing>',s)
````

````python
re.sub(r'ing',r'<ing>',s2)
````

````python
re.sub(r'(\w+ing)(\W)',r'*\g<1>*\g<2>',s)
````
