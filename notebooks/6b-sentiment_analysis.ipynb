```python
text = """How cool is that?
Well I think that is not comfortable as expected.
I hate it, it's horrible"""

blob = TextBlob(text)
```

```python
blob.polarity
```

```python
blob.sentences
```

```python
for sentence in blob.sentences:
    print(sentence, "->", sentence.sentiment.polarity)
```

```python
blob.subjectivity # is it a fact or an opinion?
```

```python
# with textblob it is easy also to get ngrams
blob.ngrams(n=3)  # getting all the trigrams
```
