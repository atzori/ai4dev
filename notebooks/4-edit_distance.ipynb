## Computing edit distance using nltk

```python
import nltk

# Definire le due stringhe da confrontare
string1 = "example"
string2 = "samples"

# Calcolare la distanza di edit utilizzando il metodo `edit_distance` di NLTK
edit_distance = nltk.edit_distance(string1, string2, substitution_cost=1.5, transpositions=False)

# Stampare la distanza di edit
print("La distanza di edit tra le due stringhe è:", edit_distance)
```
